```python
# prompt: import the following link into DataFrame: https://raw.githubusercontent.com/duchaba/book_python_genAI_genZ/blob/master/global_warming_trends/long_format_annual_surface_temp.csv

import pandas as pd

# Read the CSV file into a Pandas DataFrame
fname = 'https://raw.githubusercontent.com/duchaba/book_python_genAI_genZ/master/global_warming_trends/long_format_annual_surface_temp.csv'
df_global_warming = pd.read_csv(fname)

# Display the DataFrame
df_global_warming.head()
```

```python
# prompt: display information about df_global_warming

df_global_warming.info()
```

```python
# prompt: remove the leading "F" in the Year column in df_global_warming

df_global_warming['Year'] = df_global_warming['Year'].str[1:]
df_global_warming.head()
```

```python
# prompt: convert the year column in df_global_warming into an integer

df_global_warming['Year'] = df_global_warming['Year'].astype('int64')
df_global_warming.head()
```

```python
# prompt: plot a line graph the temperature of for the year 1961 and the year 2022 for each ISO2 from df_global_warming

df_1961 = df_global_warming[df_global_warming['Year'] == 1961]
df_2022 = df_global_warming[df_global_warming['Year'] == 2022]

import matplotlib.pyplot as plt

plt.plot(df_1961['ISO2'], df_1961['Temperature'])
plt.plot(df_2022['ISO2'], df_2022['Temperature'])

plt.xlabel('ISO2')
plt.ylabel('Average Yearly Temperature (C)')
plt.title('Temperature for 1961 and 2022')
plt.legend(['1961', '2022'])
plt.show()
```

```python
# prompt: plot a line graph the temperature for ISO2 is US for each year from df_global_warming

df_us = df_global_warming[df_global_warming['ISO2'] == 'US']
plt.plot(df_us['Year'], df_us['Temperature'])
plt.xlabel('Year')
plt.ylabel('Average Yearly Temperature (C)')
plt.title('Temperature for the US from 1961 to 2022')
plt.show()
```

```python
# prompt: plot a area graph the temperature for ISO2 is US for each year from df_global_warming

df_us = df_global_warming[df_global_warming['ISO2'] == 'US']
plt.plot(df_us['Year'], df_us['Temperature'])
plt.fill_between(df_us['Year'], df_us['Temperature'], alpha=0.2)
plt.xlabel('Year')
plt.ylabel('Average Yearly Temperature (C)')
plt.title('Temperature for the US from 1961 to 2022')
plt.show()
```

```python
# prompt: create a new dataframe from df_global_warming where average temperature group by ISO2

df_global_warming_grouped = df_global_warming.groupby('ISO2')['Temperature'].mean().reset_index()
df_global_warming_grouped.head()
```

```python
df_global_warming_grouped = df_global_warming.groupby('Country')['Temperature'].mean().reset_index()
df_global_warming_grouped.head()
```

```python
# prompt: sort df_global_warming_grouped by temperature

df_global_warming_grouped.sort_values(by='Temperature', inplace=True)
df_global_warming_grouped.head()
```

```python
# prompt: plot an area graph for temperature for first 20 record in df_global_warming_grouped and the x-axis is Country

df_global_warming_grouped.head(20).plot.area(x='Country', y='Temperature')
plt.xlabel('Country')
plt.ylabel('Average Yearly Temperature (C)')
plt.title('Temperature for the first 20 countries')
plt.show()
```

```python
# prompt: change figsize to 16,5

df_global_warming_grouped.head(20).plot.area(x='Country', y='Temperature', figsize=(16, 5))
plt.xlabel('Country')
plt.ylabel('Average Yearly Temperature (C)')
plt.title('Temperature for the first 20 countries')
plt.show()
```

```python
df_global_warming_grouped.head(20)
```

```python
# prompt: plot an area graph for temperature for last 20 record in df_global_warming_grouped and the x-axis is Country

df_global_warming_grouped.tail(20).plot.area(x='Country', y='Temperature', figsize=(16, 5))
plt.xlabel('Country')
plt.ylabel('Average Yearly Temperature (C)')
plt.title('Temperature for the last 20 countries')
plt.show()
```

```python
df_global_warming_grouped.tail(20)
```

```python
# prompt: prompt: create a new dataframe from df_global_warming where average temperature group by year

df_global_warming_by_year = df_global_warming.groupby('Year')['Temperature'].mean().reset_index()
df_global_warming_by_year.head()
```

```python
# prompt: plot bar graph the temperature for Year from df_global_warming_by_year

df_global_warming_by_year.plot.bar(x='Year', y='Temperature', figsize=(16, 4))
plt.title('Average Global Temperature by Year')
plt.xlabel('Year')
plt.ylabel('Average Temperature (C)')
plt.show()
```

```python
# prompt: create a new dataframe from df_global_warming_by_year where the Temperature is increasing by 1.0

df_global_warming_by_year_increased = df_global_warming_by_year.copy()
df_global_warming_by_year_increased['Temperature'] += 1.0
df_global_warming_by_year_increased.head()
```

```python
# prompt: plot bar graph the temperature for Year from df_global_warming_by_year_increased use color orange

df_global_warming_by_year_increased.plot.bar(x='Year', y='Temperature', figsize=(16, 4), color='orange')
plt.title('Average Global Temperature by Year (Increased by 1 Degree Celsius)')
plt.xlabel('Year')
plt.ylabel('Average Temperature (C)')
plt.show()
```

```python
# prompt: plot bar graph the temperature for Year from df_global_warming_by_year and df_global_warming_by_year_increased combine into one graph

df_combined = pd.merge(df_global_warming_by_year, df_global_warming_by_year_increased, on='Year')
df_combined.columns = ['Year', 'Temperature', 'Temperature_Increased']

# Plot the combined bar graph
df_combined.plot.bar(x='Year', y=['Temperature', 'Temperature_Increased'], figsize=(16, 4))
plt.title('Average Global Temperature by Year (Original and Increased)')
plt.xlabel('Year')
plt.ylabel('Average Temperature (C)')
plt.show()
```
