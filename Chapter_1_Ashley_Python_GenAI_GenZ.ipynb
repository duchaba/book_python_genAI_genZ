The below code is written entirely by GenAI with just the prompt:

```python
# prompt: draw a 3-d graph of z = x * exp(-(x^2 + y^2))

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x and y values
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Calculate z values
z = X * np.exp(-(X**2 + Y**2))

# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, z, cmap='viridis')

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Surface Plot of z = x * exp(-(x^2 + y^2))')

# Show the plot
plt.show()
```
